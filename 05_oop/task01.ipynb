```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            # 指定された金額を残高に追加
            self.__balance += amount
        else:
            print("エラー: 正しい金額を入力してください。")
    
    def withdraw(self, amount):
        if amount > 0 and self.__balance > amount:
            # 指定された金額を残高から引き出す
            self.__balance -= amount
        else:
            print("エラー: 残高不足または無効な金額です。")
    
    def get_balance(self):
        return self.__balance


# 呼び出し側の実装
print(f"{'-'*10} 呼び出し側の仕様で実装した場合 {'-'*10}")
bank_account = BankAccount(1000)
bank_account.deposit(500)
bank_account.withdraw(200)
print(bank_account.get_balance())
bank_account.withdraw(1500)

print(f"\n{'-'*10} 無効な金額を指定した場合 {'-'*10}")
bank_account.deposit(-500)

# print(f"\n{'-'*10} 残高に直接アクセスしようとした場合 {'-'*10}")
# print(bank_account.__balance)  # AttributeError が表示される
```
