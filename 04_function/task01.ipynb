```python
def validate_username(username):
    match username:
        # 文字数のバリデーション
        case username if len(username) < 5 or len(username) > 15:
            print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
            return False
        # 英数字のバリデーション
        case username if not username.isalnum():
            print("エラー: ユーザー名は英数字のみ使用できます")
            return False
        # 空白のバリデーション
        case username if " " in username:
            print("エラー: ユーザー名に空白を含めることはできません")
            return False
        # 上記のバリデーションを以外（正常系）
        case _:
            return True


# ユーザ名の入力
usernames = [
    "JohnDoe123",        # 正常系
    "JD",                # 異常系: 4文字以下
    "JohnDoeJohnDoe123", # 異常系: 16文字以上
    "John@Doe",          # 異常系: アルファベットと数字以外を含む
    "John Doe"           # 異常系: 空白を含む
]

for username in usernames:
    print(f"ユーザー名を {username} と入力した場合")
    # ユーザー名バリデーションの関数を呼び出す
    result = validate_username(username)
    # バリデーションの結果判定
    if result:
        print("ユーザー名は有効です\n")
    else:
        print("不正なユーザー名です\n")
```
