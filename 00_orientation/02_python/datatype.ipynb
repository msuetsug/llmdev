```python
# 整数の代入
a = 10
b = -5
c = 0

# 整数同士の計算
sum_result = a + b  # 10 + (-5) = 5
product_result = a * b  # 10 * (-5) = -50

print("a:", a)
print("b:", b)
print("c:", c)
print("合計:", sum_result)
print("積:", product_result)
```

```python
# 算術演算子の例
a = 10
b = 3

sum_result = a + b
diff_result = a - b
product_result = a * b
quotient_result = a / b    # 割り算（結果は浮動小数点）
int_quotient_result = a // b  # 整数除算
remainder_result = a % b
power_result = a ** b      # 累乗

print("加算: ", sum_result)          # 13
print("減算: ", diff_result)         # 7
print("乗算: ", product_result)      # 30
print("割算（浮動小数点）: ", quotient_result)  # 3.333...
print("整数除算: ", int_quotient_result)       # 3
print("余り: ", remainder_result)     # 1
print("累乗: ", power_result)         # 1000
```

```python
# 複合代入演算子の例
x = 10
x += 5  # x = x + 5
print("x += 5: ", x)  # 15

x -= 3  # x = x - 3
print("x -= 3: ", x)  # 12

x *= 2  # x = x * 2
print("x *= 2: ", x)  # 24

x /= 4  # x = x / 4
print("x /= 4: ", x)  # 6.0

x //= 5  # x = x // 5
print("x //= 5: ", x)  # 1.0

x %= 3  # x = x % 3
print("x %= 3: ", x)  # 1.0

x **= 3  # x = x ** 3
print("x **= 3: ", x)  # 1.0
```

```python
# 比較演算子の例
a = 10
b = 5

print("a > b: ", a > b)    # True
print("a < b: ", a < b)    # False
print("a >= b: ", a >= b)  # True
print("a <= b: ", a <= b)  # False
print("a == b: ", a == b)  # False
print("a != b: ", a != b)  # True
```

```python
# 論理演算子の例
x = 15
print("10 < x and x < 20: ", 10 < x and x < 20)  # True
print("x < 10 or x > 20: ", x < 10 or x > 20)    # False
print("not (x == 15): ", not (x == 15))          # False
```

```python
# ビット演算子の例
a = 0b1100  # 12
b = 0b1010  # 10

print("a & b: ", bin(a & b))   # 0b1000
print("a | b: ", bin(a | b))   # 0b1110
print("a ^ b: ", bin(a ^ b))   # 0b0110
print("a << 1: ", bin(a << 1)) # 0b11000
print("a >> 1: ", bin(a >> 1)) # 0b0110
```

```python
# 各データ型に対する演算子の使用例
a = 10   # int
b = 3.5  # float
c = True  # bool
d = "hello"  # string
e = [1, 2, 3]  # list

# 算術演算子
print(a + 5)   # 整数の加算 15
print(b * 2)   # 浮動小数点の乗算 7.0
print(d + " world")  # 文字列の連結 hello world
print(e * 2)   # リストの繰り返し [1, 2, 3, 1, 2, 3]

# 比較演算子
print(a > 5)   # True
print(d == "hello")  # True
print(c == True)   # True

# 論理演算子
print(c and False)  # False
print(not c)   # False

# メンバーシップ演算子
print(2 in e)  # True
print("o" in d)  # True

# 同一性演算子
f = e
print(f is e)  # True
```
