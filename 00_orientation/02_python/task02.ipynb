```python
# 商品価格を入力
price_excluding_tax = int(input("商品価格を入力してください。"))

# 商品価格から税込価格を算出
price_including_tax = int(price_excluding_tax * 1.1)

# 税込価格から送料を算出
if price_including_tax >= 2000:  # 税込価格が2000円以上の場合は送料無料
    shipping = 0
    print("送料は無料です")
else:  # 税込価格が2000円未満の場合は350円
    shipping = 350
    print(f"送料は{shipping}円です")

# 税込価格と送料の合計金額を算出
total_price = price_including_tax + shipping
print(f"送料込税込価格は{total_price}円です")
```
