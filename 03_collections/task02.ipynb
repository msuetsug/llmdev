```python
# 英単語を格納するリストを作成する
english_words_list = []
# アルファベットごとの数を格納する辞書を作成する
ascii_lowercase_keys = list("abcdefghijklmnopqrstuvwxyz")
ascii_dict = dict.fromkeys(ascii_lowercase_keys, 0)

# 英単語の入力の受付を繰り返す
while True:
    # キーボードから英単語の入力を受け付ける
    english_words = str(input("英単語を入力してください:"))
    # 入力された文字が "" なら繰り返しを抜ける
    if english_words == "":
        break
    # 英単語をリストに格納する
    english_words_list.append(english_words)

    # 英単語を構成する文字を1文字ずつ確認して、登場したアルファベットの数に1を加算する
    for word in list(english_words):
        ascii_dict[word] += 1

# 英単語のリストをABC順に並べ替える
sorted_english_words_list = sorted(english_words_list)
# 英単語のリストを一覧表示する
print(f"入力した英単語： {sorted_english_words_list}")
# アルファベットごとの数を格納する辞書から、数が0より大きい文字について「○が△個ありました」と表示する
for key, value in ascii_dict.items():
    if value > 0:
        print(f"{key} が {value} 個ありました")
```
