```python
# 学生の成績データ
grades_list = [
    [85, 90, 78, 92, 88],
    [70, 75, 80, 85, 90]
]

for grades in grades_list:
    print(f"成績リストが {grades} の場合")
    
    # 合計をリストの件数で割ることで平均を算出
    total = 0
    for score in grades:
        total += score
    average = total / len(grades)
    print(f"平均点: {average}")

    # ソートしたリストの最初と最後から最高点と最低点の検出
    sorted_grades = sorted(grades)
    print(f"最高点: {sorted_grades[-1]}")
    print(f"最低点: {sorted_grades[0]}")

    # 平均以上の成績だけをリスト化し、件数からカウントを計算
    above_average_grades = [score for score in grades if score >= average]
    print(f"平均以上の学生数: {len(above_average_grades)}\n")
```
